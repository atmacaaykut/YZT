```python
df=df.drop(['Daily Summary'],axis=1)
```

```python
df['Precip Type'].fillna('no rain', inplace=True)
# 'weather' sütununu dummy (binary) sütunlara çevir
precip_type = pd.get_dummies(df['Precip Type'], prefix='Precip_Type')
summary= pd.get_dummies(df['Summary'], prefix='Summary_Enc')
# Sonucu orijinal DataFrame'e ekleyerek göster
df = pd.concat([df, precip_type], axis=1)
df = pd.concat([df, summary], axis=1)
```

```python
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc=True)

# Mevsimlere göre gruplandır ve yeni 'Season' sütununu ekle
df['Season'] = df['Formatted Date'].dt.to_period("Q-DEC").dt.quarter.map({1: 'Winter', 2: 'Spring', 3: 'Summer', 4: 'Fall'})
season=pd.get_dummies(df['Season'], prefix='Season_Enc')
df = pd.concat([df, season], axis=1)
```

```python
df=df.drop(['Precip Type','Formatted Date','Season'],axis=1)
```

```python
df=df.drop(['Summary'],axis=1)
```
